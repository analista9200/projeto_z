### Camada Gold - Análise de uso de redes sociais pela geração Z (Brasil) - PySpark / Databrinks

```python
# Importando Funções Pyspark
from pyspark.sql.functions import *

#=======================#
# Objetivo das Análises #
#=====================#
# Metadado - Brasil #
#===================#

#1. Quantidade de horas por países e plataformas?
#2. Incidência entre gênero masculino e feminino por países e geral e ranking das idades?
#3. Tempo médio geral e por países e também relação a tempo noturno?
```

```python
# Carregando Tabela Silver
df_gold = spark.table("projeto_profissional_genz.genz.genz_social_media_silver")
```

```python
# Visualizando tabela gold.

df_gold.printSchema()
```

```python
# Pergunta `1` = Quantidade de horas por países e plataformas? (Somente Brasil)

gold_horas_plataforma_brasil = (
    df_gold.filter(col("pais") == "Brazil")
    .groupBy("pais", "plataforma_primaria")
    .agg(round(sum("uso_diario"), 2).alias("horas"))
    .orderBy(col("horas").desc())
)

print ("Quantidade de horas por País e Plataforma (Brasil):")
gold_horas_plataforma_brasil.show(10, truncate=False)
```

```python
# Salvando Tabela Gold

gold_horas_plataforma_brasil.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_horas_plataforma_brasil")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_horas_plataforma_brasil
```

```python
# Pergunta `2` = Incidência entre gênero masculino e feminino por países e geral e ranking das idades.

gold_country_count_brasil = (
    df_gold.filter(col("pais") == "Brazil")
    .groupBy("pais", "genero", "idade")
    .agg(count("idade").alias("quantidade"))
    .orderBy(col("quantidade").desc())
)

print ("Incidência entre gênero masculino e feminino no Brasil:")
gold_country_count_brasil.show (10, truncate=False)
```

```python
# Salvando Tabela Gold

gold_country_count_brasil.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_country_count_brasil")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_country_count_brasil
```

```python
#3. Tempo médio geral e por países e também relação a tempo noturno? (Brasil)

gold_country_time_brasil = (
    df_gold.filter(col("pais") == "Brazil")
    .groupBy("pais", "tempo_medio_sessao", "uso_noturno")
    .agg(avg("idade").alias("idade"))
    .orderBy(col("tempo_medio_sessao").desc())
)

print ("Tempo Médio no Brasil, uso noturno e idade:")
gold_country_time_brasil.show(10, truncate=False)
```

```python
# Salvando Tabela Gold

gold_country_time_brasil.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_country_time_brasil")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_country_time_brasil
```
