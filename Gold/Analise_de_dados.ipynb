### Camada Gold - Análise de uso de redes sociais pela geração Z - PySpark / Databrinks

```python
# Importando Funções Pyspark
from pyspark.sql.functions import *

#=======================#
# Objetivo das Análises #
#=======================#

#1. Quantidade de horas por países e plataformas?
#2. Incidência entre gênero masculino e feminino por países e geral e ranking das idades?
#3. Tempo médio geral e por países e também relação a tempo noturno?
```

```python
# Carregando Tabela Silver
df_gold = spark.table("projeto_profissional_genz.genz.genz_social_media_silver")
```

```python
# Visualizando tabela gold.

df_gold.printSchema()
```

```python
# Pergunta `1` = Quantidade de horas por países e plataformas?

gold_horas_plataforma = (
    df_gold.groupBy("pais", "plataforma_primaria")
    .agg(round(sum("uso_diario"), 2).alias("horas"))
    .orderBy(col("horas").desc())
)

print ("Quantidade de horas por País e Plataforma:")
display(gold_horas_plataforma.limit(10))
```

```python
# Salvando Tabela Gold

gold_horas_plataforma.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_horas_plataforma")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_horas_plataforma
```

```python
# Pergunta `2` = Incidência entre gênero masculino e feminino por países e geral e ranking das idades.

gold_country_count = (
    df_gold.groupBy("pais", "genero", "idade")
    .agg(count("idade").alias("quantidade"))
    .orderBy(col("quantidade").desc())
)

gold_main_count = (
    df_gold.groupBy("genero")
    .agg(count("idade").alias("quantidade"))
    .orderBy(col("quantidade").desc())
)

print ("Incidência entre gênero masculino e feminino por país:")
gold_country_count.show (10, truncate=False)
print ("Incidência entre gênero masculino e feminino geral:")
gold_main_count.show(3, truncate=False)
```

```python
# Salvando Tabela Gold

gold_country_count.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_country_count")

gold_main_count.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("projeto_profissional_genz.genz.gold_main_count")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_country_count
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_main_count
```

```python
#3. Tempo médio geral e por países e também relação a tempo noturno?

gold_country_time = (
    df_gold.groupBy("pais", "tempo_medio_sessao", "uso_noturno")
    .agg(avg("idade").alias("idade"))
    .orderBy(col("tempo_medio_sessao").desc())
)

print ("Tempo Médio por Países, uso noturno e idade:")
gold_country_time.show(10, truncate=False)
```

```python
# Salvando Tabela Gold

gold_country_time.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("projeto_profissional_genz.genz.gold_country_time")
```

```python
%sql
SELECT * FROM projeto_profissional_genz.genz.gold_country_time
```
