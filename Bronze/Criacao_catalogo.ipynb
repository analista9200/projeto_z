```sql
--Criação de Catalogo--
CREATE CATALOG IF NOT EXISTS projeto_profissional_genz;
```

```sql
--Criando Schema--
CREATE SCHEMA IF NOT EXISTS projeto_profissional_genz.genz;
```

```sql
CREATE VOLUME IF NOT EXISTS projeto_profissional_genz.genz.dados;
```

/Volumes/projeto_profissional_genz/genz/dados

```sql
CREATE OR REPLACE TEMP VIEW vw_csv_gensocial
USING CSV
OPTIONS (
  path "/Volumes/projeto_profissional_genz/genz/dados/genz_social_media.csv",
  header "true",
  inferSchema "true",
  sep ","
);
```

```sql
CREATE OR REPLACE TABLE projeto_profissional_genz.genz.genz_social_media_bronze
USING DELTA
AS
SELECT * FROM vw_csv_gensocial;
```
